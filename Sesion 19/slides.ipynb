### Nota: 
Como las primeras dos filas no contienen información necesario , las eliminaremos 

```python
data = pd.read_csv('data.csv', header = 0, skiprows= range(1,3))
```

```python
data.shape
```

```python
data.head()
```

```python
'''hacemos la transpuestas del dataset, esto 
    es intercambiar filas <->columnas'''

temp = data.T
temp
```

Tomando la primera fila como la cabecera del dataset

`.iloc[posicion]` -> con esta propiedad podemos trabjar con las posiciones de lo que seleccionamos. 

```python
new_header = temp.iloc[0]
new_header
```

```python
temp = temp[1:] #slicing del dataset
temp
```

Renombrando las columnas del temp

```python
temp.columns = new_header
temp
```

```python
temp.head(2)
```

```python
temp.head(3)
```

## Crear un nuevo dataframe  `Pais vs Obesidad`

```python
df_country_obesity = pd.DataFrame(columns=['Country','Obesity%'])
df_country_obesity
```

```python
for i , j in temp[temp.columns[1:]].items():
    df_country_obesity = df_country_obesity.append(
            pd.DataFrame({'Country':i,'Obesity%': j.values}))
```

```python

df_country_obesity
```

## Crear un nuevo dataframe  `Year vs  Sex`

```python
df_year_sex = pd.DataFrame(columns = ['Year', 'Sex'])
df_year_sex
```

```python
for year in range(2016,1974,-1):
    df_year_sex = df_year_sex.append(
        pd.DataFrame({'Year': year ,
                      'Sex':['Both Sexes', 'Male', 'Female'] })
                    )
```

```python
df_year_sex
```

### Unir ambos dataframe con `.join`

```python
df_year_sex.reset_index( drop = True, inplace = True)
df_year_sex
```

```python
data_frame = df_country_obesity.join(df_year_sex, how = 'left')
data_frame
```

```python
data_frame.sort_values(by = ['Country','Year'], inplace = True)
data_frame
```

```python
# Crear un nievo Data Frame con la informacion de El Salvador

sv_data = data_frame[data_frame['Country'] == 'El Salvador']

sv_data
```

## Limpiando los datos que no eran numéricos para poder hacer análisis

```python
sv_data['Obesity%']= sv_data['Obesity%'].str.extract(r'(\d+\.\d+)').astype(float)
```

```python
sv_data
sv_data.to_csv('sv_data.csv')
```

### VErificando si quedaron valores nulos

```python
sv_data.isnull().sum()
```

### Gráfico de Obesidad promedio por año en SV

```python
import matplotlib.pyplot as plt
```

```python
average_by_year = sv_data.groupby('Year')['Obesity%'].mean()
average_by_year.round(1)
```

```python
plt.figure(figsize=(10,6))
average_by_year.plot(kind = 'bar')
plt.ylabel('Promedio del % de Obesidad')
plt.grid()
plt.tight_layout()
```

```python
plt.figure(figsize=(10,6))
plt.plot(sv_data['Year'],sv_data['Obesity%'], 
         marker = 'o' ,linestyle = '-', color = 'red')
plt.ylabel('Promedio del % de Obesidad')
plt.grid()
plt.tight_layout()
```
