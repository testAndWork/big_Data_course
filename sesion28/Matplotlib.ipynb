## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#%matplotlib notebook
```

`np.linespace(inicio, fin, numero de particiones)`

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--' )
```

```python
fig.savefig('figure.png')
```

## Subplots

```python

fig = plt.figure()

plt.subplot(2,1,1) #( numero de filas ,columnas, numero de imagen )
plt.plot(x, np.sin(x), '-')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), '--' )
```

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x), '-')
ax[1].plot(x, np.cos(x), '--' )
```
